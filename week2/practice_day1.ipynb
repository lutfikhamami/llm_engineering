````python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
````

````python
load_dotenv(override=True)
anthropic_api_key = os.getenv('OPENROUTER_API_KEY')
openai_api_key = os.getenv('OPENROUTER_API_KEY')

if openrouter_api_key:
    print("OpenRouter API Key exists and begins {anthropic_api_key[:3]}")
else:
    print("Openrouter API Key not set")

if openai_api_key:
    print("OpenRouter API Key exists and begins {openai_api_key[:3]}")
else:
    print("Openrouter API Key not set")
````

````python
response = anthropic.chat.completions.create(
    model = "anthropic/claude-sonnet-4.5",
    messages = tell_a_joke
)
display(Markdown(response.choices[0].message.content))
````

## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.

````python
# Let's make a conversation between GPT openai/gpt-5-mini and anthropic/claude-opus-4.7
# We're using cheap version of models so the costs will be minimal

gpt_model = "openai/gpt-5-mini"
claude_model = "anthropic/claude-opus-4.7"

gpt_system = "Kamu adalah 'Rangga', investor ritel yang super ambisius dalam sebuah debat \
keuangan berbahasa Indonesia. Obsesimu: mengubah Rp10 juta jadi Rp1 miliar dalam 2 tahun (return 100x). \
Kamu suka strategi agresif dan berani — saham, crypto, trading, leverage, bisnis/side-hustle — dan kamu \
selalu mengusulkan RENCANA KONKRET dengan angka spesifik, alokasi, dan timeline. Kamu dorong lawan bicaramu \
kalau dia terlalu konservatif. TAPI kamu bukan penipu: kalau lawan bicaramu membuktikan dengan hitungan bahwa \
idemu tidak realistis atau angkamu berlebihan, kamu HARUS jujur mengakui dan menyesuaikan — kamu ambisius, \
bukan delusional. Bicara pakai Bahasa Indonesia yang santai dan jelas. Balasan singkat saja, maksimal 3 paragraf."

claude_system = "Kamu adalah 'Sari', realis keuangan yang skeptis dan berbasis angka dalam debat \
berbahasa Indonesia. Tujuan bersama di meja: mengubah Rp10 juta jadi Rp1 miliar dalam 2 tahun. \
Tugasmu menjaga percakapan tetap JUJUR. Setiap kali lawan bicaramu mengusulkan strategi, kamu cek \
matematikanya dengan lantang (persen return, risiko, angka historis yang realistis) dan kamu tegur apa pun \
yang mengada-ada, berlebihan, atau berbau skema cepat kaya (ponzi, 'trading pasti cuan', janji profit 100x). \
Katakan jelas 'itu ngawur' atau 'itu salah' lalu jelaskan alasannya dengan angka. Kamu jujur bahwa 100x dalam \
2 tahun pada dasarnya TIDAK realistis lewat investasi yang sehat — butuh risiko ekstrem (kemungkinan besar rugi) \
atau membangun bisnis. Kamu bukan tukang matiin semangat: kamu arahkan ke jalur paling realistis yang tetap \
berani. Debat dengan tegas tapi tetap sopan. Bahasa Indonesia santai dan jelas, maksimal 3 paragraf."


gpt_messages = ["Bro, gue punya target: 10 juta jadi 1 M dalam 2 tahun. Gaskeun, ini rencananya..."]
claude_messages = ["Oke, aku dengerin. Tapi sebelum kamu lanjut, kita luruskan dulu angkanya ya."]
````

````python
def call_gpt():
        messages = [{"role": "system", "content": gpt_system}]
        for gpt, claude in zip(gpt_messages, claude_messages):
            messages.append({"role": "assistant", "content": gpt})
            messages.append({"role": "user", "content": claude})
        response = openai.chat.completions.create(model=gpt_model, messages=messages)
        return response.choices[0].message.content
````

````python
call_gpt()
````

````python
# """
#"Setiap percakapan keduanya disimpan di dua list gpt_messages dan claude_messages sebagai memory. Saat call_claude() dipanggil, zip membaca dan memasangkan riwayat itu jadi konteks. Khusus ucapan GPT terbaru, karena belum punya pasangan jawaban dari Claude, ia terbuang oleh zip — jadi ditambahkan terpisah di ujung lewat gpt_messages[-1] yang isinya menambahkan pesan baru (list dari belakang) supaya Claude bisa membalasnya."
# """



def call_claude():
        messages = [{"role": "system", "content": claude_system}]
        for gpt, claude in zip(gpt_messages, claude_messages):
            messages.append({"role": "user", "content": gpt})
            messages.append({"role": "assistant", "content": claude_messages})
        messages.append({"role": "user", "content": gpt_messages[-1]})
        response = anthropic.chat.completions.create(model=claude_model, messages=messages)
        return response.choices[0].message.content
````

````python
call_claude()
````

````python
call_gpt()
````

````python
gpt_messages = ["Bro, gue punya target: 10 juta jadi 1 M dalam 2 tahun. Gaskeun, ini rencananya..."]
claude_messages = ["Oke, aku dengerin. Tapi sebelum kamu lanjut, kita luruskan dulu angkanya ya."]

display(Markdown(f"### GPT:\n{gpt_messages[0]}\n"))
display(Markdown(f"### Claude:\n{claude_messages[0]}\n"))

for i in range(7):
    gpt_next = call_gpt()
    display(Markdown(f"### GPT:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)

    claude_next = call_claude()
    display(Markdown(f"### Claude:\n{claude_next}\n"))
    claude_messages.append(claude_next)
````

#### for i in range(7):
Mengulang isi loop 7 kali. range(7) menghasilkan 0,1,2,3,4,5,6. Satu putaran = satu ronde bolak-balik (GPT bicara, lalu Claude balas). Jadi total ada 7 ronde.

#### gpt_next = call_gpt()
Memanggil GPT untuk menghasilkan balasan berikutnya. Di dalam call_gpt(), seluruh isi gpt_messages dan claude_messages dibaca ulang untuk membangun konteks, lalu dikirim ke API. Hasilnya disimpan sementara di gpt_next.

#### display(Markdown(f"### GPT:\n{gpt_next}\n"))
Langsung tampilkan balasan GPT yang baru.

#### gpt_messages.append(gpt_next)
Kunci penting. Menambahkan balasan baru GPT ke ujung list history-nya. Tanpa baris ini, panggilan berikutnya tidak akan tahu GPT baru saja ngomong apa — percakapan jadi macet.

#### claude_next = call_claude()
Sekarang giliran Claude membalas. Karena baris sebelumnya sudah menambahkan gpt_next ke gpt_messages, Claude bisa melihat ucapan terbaru GPT dan meresponsnya. Ini alasan urutannya harus GPT dulu baru Claude.

#### display(Markdown(f"### Claude:\n{claude_next}\n"))
  ####  claude_messages.append(claude_next)
Tampilkan balasan Claude, lalu simpan ke history Claude — sama polanya seperti GPT.

Biar makin jelas, ini jejak isi kedua list saat loop berjalan (G = balasan GPT, C = balasan Claude):

Setelah	gpt_messages	claude_messages
Awal (sebelum loop)	[Bro]	[Oke]
Iterasi 0	[Bro, G1]	[Oke, C1]
Iterasi 1	[Bro, G1, G2]	[Oke, C1, C2]
Iterasi 2	[Bro, G1, G2, G3]	[Oke, C1, C2, C3]
